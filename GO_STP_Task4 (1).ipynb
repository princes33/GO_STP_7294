16.How to Accessing/Changing specific elements, rows, columns, etc in Numpy array? Example -

```python
p= np.array([[ 10,20,30,40,50,60,80],[70,90,101,111,122,113,144]])
print("Accessing Element 13 : ",p[1,5])
print("1st Row: ",p[0])
print("3rd Column:",p[0,2])
print(p[0,1:6:2])
p[1,5] = 20
print("Replace Element 13 : ",p[1,5])
```

18. Create the following pattern without hardcoding. Use only numpy functions and the below input array a.
Input: a = np.array([1,2,3])
Desired Output: array([1, 1, 1, 2, 2, 2, 3, 3, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3])


```python

a = np.array([1,2,3])
print(a)

a = np.append(np.repeat(a,3),np.tile(a,3))
print(a)
```
