3. Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average speed
in 8 days.
days=[1,2,3,4,5,6,7,8]
max_speed=[80,91,92,88,77,79,76,75]
min_speed=[42,43,40,42,33,36,34,35]
avg_speed=[46,58,57,56,40,42,41,36]

```python
days=np.array([1,2,3,4,5,6,7,8])
max_speed=np.array([80,91,92,88,77,79,76,75])
min_speed=np.array([42,43,40,42,33,36,34,35])
avg_speed=np.array([46,58,57,56,40,42,41,36])
fig,(ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize = (10,10))
ax1.plot(days,max_speed,'--', linewidth=4)
ax1.scatter(days,max_speed,color='yellow',marker='p',linewidth=9)
ax1.set_title('Days vs Maximum Speed')
ax1.set_xlabel('Days')
fig,ax1.set_ylabel('Max Speed')
ax2.plot(days,min_speed, '-r', linewidth=3)
ax2.scatter(days,min_speed,color='green',marker='D',linewidth=12)
ax2.set_title('Days vs Minimum Speed')
ax2.set_xlabel('Days')
ax2.set_ylabel('Min Speed')
ax3.plot(days,avg_speed, '.-m', linewidth=6)
ax3.scatter(days,avg_speed,color='c',marker='x',linewidth=20)
ax3.set_title('Days vs Average Speed')
ax3.set_xlabel('Days')
ax2.set_ylabel('Average Speed')
fig.suptitle('Speed Plot', y = 1.00, fontsize=20)
fig.tight_layout()
```

4. Plotting a basic sine graph by adding more features. Adding Multiple plots by
Superimposition like cosine wave.


```python

x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
plt.figure(figsize=(12,3))
plt.plot(x,y,'b', linewidth = 3)
plt.plot(x,z,'--g',linewidth = 3)
plt.xlabel("Angle ")
plt.ylabel("Magnitude")
plt.title(" Superimposition Sine/Cosine curve")
plt.legend(['sine','cosine'])
plt.show()
```

5. Plot Simple bar chart showing popularity of Programming Languages.
Languages = ['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]
Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal
bar chart using barh function.

```python

Languages = ['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = np.array([56, 39, 34, 34, 29])
Security = np.array([44 ,36 ,55, 50, 42])
fig,(ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize = (10,10))
ax1.barh(Languages,Popularity,color= 'c')
ax1.set_title('Horizontal Bar plot of Popularity')
ax1.set_ylabel('Popularity')
ax2.bar(Languages,Security,color= 'm',edgecolor='black' )
ax2.set_title('Bar plot of Security')
ax2.set_ylabel('Security')
ax3.hist(Security,color= 'y',edgecolor='b' )
ax3.set_title('Histogram of Security')
ax3.set_xlabel('Bins')
ax3.set_ylabel('security')
fig.suptitle('Programming Languages', y = 1.05, fontsize=20)
```
